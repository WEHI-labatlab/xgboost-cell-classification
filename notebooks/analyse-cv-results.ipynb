```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
model = pickle.load(open("/Volumes/yokote.k/xgboost/resources/output/NSCLCcohort_FM_train_CD16_None_all/bayes_cv_model.sav", 'rb'))
```

```python
model.best_params_
```

```python
model.cv_results_
```

```python
top_k = 50

f_i = list(zip(model.feature_names_in_,model.feature_importances_))
f_i.sort(key = lambda x : x[1])

f_i_k = f_i[len(f_i) - top_k:]


figure(figsize=(20, 20), dpi=80)
plt.barh([x[0] for x in f_i_k],[x[1] for x in f_i_k])
# plt.xscale('log')
plt.show()
```

```python
len(f_i_k)
```

```python
bottom_k = 200

f_i = list(zip(model.feature_names_in_,model.feature_importances_))
f_i.sort(key = lambda x : x[1])

f_i_k = f_i[:bottom_k+1]


figure(figsize=(20, 20), dpi=80)
plt.barh([x[0] for x in f_i_k],[x[1] for x in f_i_k])

plt.show()
```
